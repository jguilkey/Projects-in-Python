```python
from urllib.request import urlopen
import json
import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "svg"     #Used to render interactive maps on GitHub
```

```python
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)
```

```python
df = pd.read_csv('us-counties.csv',
                   dtype={'fips': str})

df = df[df['date']=='2020-04-15']
```

```python
fig = px.choropleth(df, geojson=counties, locations='fips', color='cases',
                           color_continuous_scale="Viridis",
                           range_color=(0, 1000),
                           scope="usa",
                           hover_name = 'county',
                           labels = {'cases':'Covid Cases'},
                           title = 'Covid Cases in US by County (04-15-2020)')

fig.show()     #Interactive choropleth map 
```
